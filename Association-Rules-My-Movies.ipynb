```python
import pandas as pd
import numpy as np
import matplotlib. pyplot as plt
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
movie=pd.read_csv("D:/Assignments Excel R/assignment 9/my_movies.csv")
movie
```

```python
movie_df=movie.iloc[:,5:15]
```

```python
movie_df
```

 ## Apriori Algorithm

```python
# Apriori(df, min_support=0.4, use_colnames=Ture)
```

```python
frequent_itemsets=apriori(movie_df, min_support=0.1, use_colnames=True)
frequent_itemsets['length']=frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

 ## Association Rules

```python
rules=association_rules(frequent_itemsets, min_threshold=1)
rules
```

```python
rules[(rules['lift']>=2.5)&
     (rules['confidence']>=1)]
```

 ## Visualization

```python
import networkx as nx
```

 ## NetworkX Visualization

```python
fig, ax=plt.subplots(figsize=(15,10))
GA=nx.from_pandas_edgelist(rules,source='antecedents',target='consequents')
nx.draw(GA,with_labels=True)
plt.show()
```

 ## Support vs Confidence

```python
plt.scatter(rules['support'], rules['confidence'], alpha=1)
plt.xlabel('support')
plt.ylabel('confidence')
plt.title('Support vs Confidence')
plt.show()
```

 ## Support vs Lift

```python
plt.scatter(rules['support'], rules['lift'], alpha=1)
plt.xlabel('support')
plt.ylabel('lift')
plt.title('Support vs Lift')
plt.show
```

 ## Lift vs Confidence

```python
fit = np.polyfit(rules['lift'], rules['confidence'], 1)
fit_fn = np.poly1d(fit)
plt.plot(rules['lift'], rules['confidence'], 'yo', rules['lift'], 
 fit_fn(rules['lift']))
plt.title('Lift vs Confidence')
```
